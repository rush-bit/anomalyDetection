```python
import pandas as pd

df = pd.read_csv('clustered_features_with_anomalies.csv')

# Top anomalies per cluster (EBM residual score)
top_ebm = (df.sort_values(['cluster','anomaly_score_ebm'], ascending=[True, False]).groupby('cluster').head(10))
top_ebm[['deviceID','tripID','cluster','anomaly_score_ebm','alert_ebm','anomaly_score_iforest','alert_iforest']].head(20)
```

```python
agree = df[(df['alert_ebm'] == True) & (df['alert_iforest'] == True)]
agree = agree.sort_values('anomaly_score_ebm', ascending=False)
agree[['deviceID','tripID','cluster','anomaly_score_ebm','anomaly_score_iforest']].head(20)
```
