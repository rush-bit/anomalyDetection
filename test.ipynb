### This file contains the segregation of testset and trainset from the dataset

```python
import pandas as pd
```

```python
df=pd.read_csv('fleet_.csv',low_memory=False)
```

```python
# 1) Sample exactly 1,020,510 rows reproducibly
test = df.sample(n=1_020_510, random_state=42)

# 2) Remove the sampled rows to get the remainder
train_pool = df.drop(index=test.index)

# (optional) Reset indices
test = test.reset_index(drop=True)
train_pool = train_pool.reset_index(drop=True)
```
